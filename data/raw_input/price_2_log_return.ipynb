```python
import pandas as pd
import numpy as np
```

```python
price = pd.read_csv('r3000_price.csv')
price.shape
price.head()
```

```python
x = np.log(price.iloc[:,1:])
x.shape
x.head()
```

```python
y = x.diff()
y.shape
y.head()
```

```python
ret = pd.concat([price[['Date']], y], axis=1)
ret.head()
```

```python
ret.to_csv('r3000_log_return.csv')
```
