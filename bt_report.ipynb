# Backtest result of 4-factor model
Baruch Investment Club  
04-19-2017

Basic Backtest configuration:  
- Daterange: 2014-1-1 to 2016-12-31
- Rebalance: Monthly
- Factors: Vol10, momentum, beta, market cap
- Factor preprocessing: Scale and wisorize with +/- 3 standard deviation
- Fitting window: 2 years
- Trade pool filters: 5 day volume >= 300K, price > \$10, not in MA 
- Optimization: Single name has weight cap of 20%

Results are as follows, all metrics are annualized.

```python
# Result
pnl = pd.read_csv('./output/monthly rebalance pnl.csv')
pnl.date = pd.to_datetime(pnl.date, format='%Y-%m-%d')

pmfc = (cagr(pnl),\
        vol(pnl),\
        sharpe_ratio(pnl),\
        max_drawdown(pnl),\
        drawdown_length(pnl))
print('CAGR:\t\t%.4f\n'\
        'Volatility:\t%.4f\n'\
        'Sharpe_ratio:\t%.4f\n'\
        'Max drawdown:\t%.4f\n'\
        'Drawdown length: %d' % pmfc)

plot_nav(pnl)
```

If we do weekly rebalance instead

```python
# Result
pnl = pd.read_csv('./output/weekly rebalance pnl.csv')
pnl.date = pd.to_datetime(pnl.date, format='%Y-%m-%d')

pmfc = (cagr(pnl),\
        vol(pnl),\
        sharpe_ratio(pnl),\
        max_drawdown(pnl),\
        drawdown_length(pnl))
print('CAGR:\t\t%.4f\n'\
        'Volatility:\t%.4f\n'\
        'Sharpe_ratio:\t%.4f\n'\
        'Max drawdown:\t%.4f\n'\
        'Drawdown length: %d' % pmfc)

plot_nav(pnl)
```
